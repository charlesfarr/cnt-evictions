# Importing packages & setting global variables 

```python
import os
import pandas as pd
```

```python
# setting directories for data i/o

data_directory = 'C:/Users/charl/Documents/CNT 2020/Evictions/cnt-evictions/data/'
hta_directory = 'hta_blocks/'
eviction_lab_directory = 'ev_lab_blocks/'
```

```python
# updating pandas view settings

pd.set_option('display.max_columns', 999)
```

# Pulling files names for HTA & Eviction Lab datasets

```python
# setting empty lists for import
hta_data_files = []
ev_lab_files = []
```

```python
# getting list of available hta_files

with os.scandir(data_directory+hta_directory) as entries:
    for entry in entries:
#         print(entry.name)
        hta_data_files.append(entry.name)
```

```python
print(hta_data_files)
```

```python
# getting list of available ev_lab_files

with os.scandir(data_directory+eviction_lab_directory) as entries:
    for entry in entries:
        ev_lab_files.append(entry.name)
```

```python
print(ev_lab_files)
```

# Import H+T Data

```python
# Set selected states using state abbreviations

states_of_interest = ['ny','ca','il','tx','az','pa','fl','oh','nc','in','wa','co','dc']
```

```python
# Import data for the states of interest

hta_state_data = {}

for file in hta_data_files:
    for state in states_of_interest:
        if state in file:
            hta_state_data[state] = pd.read_csv(data_directory+hta_directory+file)
#         print('state: '+state)
            
```

```python
all_hta_state_data = {}

for file in hta_data_files:
    state_abbrev = file.split('_')[3]
    state_abbrev = state_abbrev.split('.')[0]
    
    print('state: '+state_abbrev)
    
    all_hta_state_data[state_abbrev] = pd.read_csv(data_directory+hta_directory+file)
    all_hta_state_data[state_abbrev]['state'] = state_abbrev
```

```python
hta_state_data['ny'].head()
```

# HTA Data Cleaning

```python
for key, df in hta_state_data.items():
    print(key)
```

```python
# clean columns in HTA data files

for key, df in hta_state_data.items():
    df['blkgrp'].replace(regex=True,inplace=True,to_replace='"',value=r'')
    df['cbsa'].replace(regex=True,inplace=True,to_replace='"',value=r'')    
```

```python
hta_state_data['ny'].head()
```

# Eviction Lab Import

```python
# Import data for the states of interest

ev_lab_state_data = {}

for file in ev_lab_files:
#     print('file: '+file)
    for state in states_of_interest:
        if state in file:
            ev_lab_state_data[state] = pd.read_csv(data_directory+eviction_lab_directory+file,low_memory=False)
            print('state imported: '+state)
            
```

```python
ev_lab_state_data['ny'].head()
```

# Eviction Lab Data Cleaning

# Join all HTA datasets into one

```python
all_hta_state_data['ny'].head()
```

```python
for key, df in all_hta_state_data.items():
    df['blkgrp'].replace(regex=True,inplace=True,to_replace='"',value=r'')
    df['cbsa'].replace(regex=True,inplace=True,to_replace='"',value=r'')
```

```python
all_hta_state_data['ny'].head()
```

```python
# Concating all HTA focus datasets together

aggregated_hta_df = pd.concat(all_hta_state_data.values(), ignore_index = True)

aggregated_hta_df.head()
```

```python
aggregated_hta_df.groupby('cbsa')['population'].sum().sort_values(ascending=False)
```

```python
cbsa_by_pop_df = aggregated_hta_df.groupby('cbsa').agg({'population':['sum'],'blkgrp':['count'],'households':['sum']}).reset_index()
```

```python
cbsa_by_pop_df.columns
```

```python
cbsa_by_pop_df.columns = ['cbsa','pop_sum','blkgrp_count','household_sum']
```

```python
cbsa_by_pop_df.sort_values(by='pop_sum',ascending=False,inplace=True)
```

```python
cbsa_by_pop_df.reset_index(drop=True,inplace=True)
```

```python
cbsa_by_pop_df.head()
```

```python
# cbsa_by_pop_df.drop(['index'],axis=1,inplace=True)
```

```python
top_20_cbsa = cbsa_by_pop_df.nlargest(20, 'pop_sum')['cbsa'].to_list()
```

```python
top_20_cbsa
```

# Select HTA blockgroup level data based on top 20 CBSA

```python
top_20_cbsa_hta_df = aggregated_hta_df.loc[aggregated_hta_df['cbsa'].isin(top_20_cbsa)].reset_index(drop=True,inplace=False)
```
