# Import H+T Data

```python
all_hta_state_data = {}

for file in hta_data_files:
    state_abbrev = file.split('_')[3]
    state_abbrev = state_abbrev.split('.')[0]
    
    print('state: '+state_abbrev)
    
    all_hta_state_data[state_abbrev] = pd.read_csv(data_directory+hta_directory+file)
    all_hta_state_data[state_abbrev]['state'] = state_abbrev
```

# Join all HTA datasets into one

```python
all_hta_state_data['ny'].head()
```

```python
for key, df in all_hta_state_data.items():
    df['blkgrp'].replace(regex=True,inplace=True,to_replace='"',value=r'')
    df['cbsa'].replace(regex=True,inplace=True,to_replace='"',value=r'')
```

```python
all_hta_state_data['ny'].head()
```

```python
# Concating all HTA focus datasets together

aggregated_hta_df = pd.concat(all_hta_state_data.values(), ignore_index = True)

aggregated_hta_df.head()
```

```python
aggregated_hta_df.groupby('cbsa')['population'].sum().sort_values(ascending=False)
```

# Identify Top 20 CBSAs by Population

```python
cbsa_by_pop_df = aggregated_hta_df.groupby('cbsa').agg({'population':['sum'],'blkgrp':['count'],'households':['sum']}).reset_index()
```

```python
cbsa_by_pop_df.columns
```

```python
cbsa_by_pop_df.columns = ['cbsa','pop_sum','blkgrp_count','household_sum']
```

```python
cbsa_by_pop_df.sort_values(by='pop_sum',ascending=False,inplace=True)
```

```python
cbsa_by_pop_df.reset_index(drop=True,inplace=True)
```

```python
cbsa_by_pop_df.head()
```

```python
# cbsa_by_pop_df.drop(['index'],axis=1,inplace=True)
```

```python
top_20_cbsa = cbsa_by_pop_df.nlargest(20, 'pop_sum')['cbsa'].to_list()
```

```python
top_20_cbsa
```

# Select HTA blockgroup level data based on top 20 CBSA

```python
top_20_cbsa_hta_df = aggregated_hta_df.loc[aggregated_hta_df['cbsa'].isin(top_20_cbsa)].reset_index(drop=True,inplace=False)
```
